# Analysis of Active Sellers Performance using Pandas

This project analyzes the performance of active sellers using a dataset.

The dataset includes seller details such as:
- Number of products listed
- Total units sold
- Ratings and reviews
- Pricing and discount information

**Steps Performed**
1. Data Loading
2. Data Cleaning (handling missing values and duplicates)
3. Feature Engineering (creating new columns)
4. Data Analysis (sorting, filtering, grouping)
5. Identifying successful sellers

**Key Insights**
- Sellers with higher ratings tend to have higher sales
- Discounts influence product sales significantly
- Sellers with more listed products generally sell more units
- Pricing strategy plays a key role in performance

This analysis helps understand the factors affecting seller success such as pricing, ratings, and product volume.
